```python
import pandas as pd

titanicdf = pd.read_csv("../../data/test.csv", index_col=0)
```

To Do:

1. Impute missing values in Age (mean, optional Age binning) and Embarked (mode)
2. One-Hot-Encode Sex, Pclass, Embarked
3. Combine info from SibSp and Parch to new OHE column named "has_relatives"
4. Scale Age and Fare (Min-Max-Scaler)

Imputations:

```python
titanicdf["Age"] = titanicdf["Age"].fillna(round(titanicdf["Age"].mean(),0))
```

```python
# optional age binning

age_bins = pd.cut(titanicdf['Age'], bins=[0, 18, 50, 85], labels=['young', 'middleaged', 'old'])  
titanicdf["AgeClass"] = age_bins
titanicdf[["is_middleaged","is_old"]] = pd.get_dummies(titanicdf["AgeClass"], drop_first=True)
```

```python
titanicdf["Embarked"] = titanicdf["Embarked"].fillna(titanicdf["Embarked"].mode()[0])
```

OHE's:

```python
titanicdf["is_Male"] = pd.get_dummies(titanicdf["Sex"], drop_first=True)
```

```python
titanicdf[["is_Pclass_2", "is_Pclass_3"]] = pd.get_dummies(titanicdf["Pclass"], drop_first=True)
```

```python
titanicdf[["emb_q","emb_s"]] = pd.get_dummies(titanicdf["Embarked"], drop_first=True)
```

Combine SibSp and Parch

```python
titanicdf["has_relatives"] = 0
```

```python
titanicdf.loc[(titanicdf["SibSp"] > 0) | (titanicdf["Parch"] > 0), "has_relatives"] = 1
```

Scale Age and Fare column

```python
age_range = titanicdf["Age"].max() - titanicdf["Age"].min()
titanicdf["scld_Age"] = (titanicdf["Age"] - titanicdf["Age"].min()) / age_range
```

```python
fare_range = titanicdf["Fare"].max() - titanicdf["Fare"].min()
titanicdf["scld_Fare"] = (titanicdf["Fare"] - titanicdf["Fare"].min()) / fare_range
```
