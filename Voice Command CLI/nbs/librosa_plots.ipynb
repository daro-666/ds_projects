```python
def plot_wav_spec(filepath_to_wav:str, hop_length=16, size=18) -> None:
    ''' creates 2in1 plot of waveform and corresponding spectrogram
    '''

    n_fft = 512
    hop_length = hop_length
    win_length = None

    y, sr = lr.load(filepath_to_wav)

    fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, figsize=(size, size*(4/9)))
    
    lr.display.waveshow(y, ax=ax1)
    
    ax1.label_outer()
    ax1.set_title('Waveform')
    ax1.set_ylabel("Amplitude")
    
    y_spec = lr.amplitude_to_db(np.abs(lr.stft(y, hop_length=hop_length, n_fft=n_fft, win_length=win_length,)), ref=np.max)
    spec = lr.display.specshow(y_spec, y_axis='linear', x_axis='time', sr=sr, hop_length=hop_length, n_fft=n_fft, win_length=win_length, ax=ax2)
    
    ax2.set_title('Spectrogram')
    ax2.set_xlabel('Time (s)')
    ax2.set_ylabel('Frequency (Hz)')
    
    fig.colorbar(spec, ax=(ax1, ax2),format="%+2.f dB")
    #plt.suptitle("Voice command: \"close\"").set_fontsize('x-large')
    #plt.savefig(f"../plots/files.pdf")
    plt.show()
```

```python
plot_wav_spec(media_fp)
```
